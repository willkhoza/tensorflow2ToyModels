```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
```

```python
data = pd.read_csv('/home/lindo/Data/NLP/Marito/st_imbd')
X = data['st']
y = data['sentiment']
```

```python
X_tokens = tokenizer.texts_to_sequences(X)
```

```python
atokenizer = dict(map(reversed, tokenizer.word_index.items()))
```

```python
y = y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tokens, y, test_size=0.33, random_state=42)
```

```python
X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train, maxlen = 20, padding = 'pre', truncating = 'pre')
```

```python
def atoken(seq):
    out = []
    for key in seq:
        if key >0:
            out.append(atokenizer[key])
    return out
```

```python
atoken(X_train[1])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 64),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='tanh')
])
```

```python
model.compile(loss='mse',
              optimizer=tf.keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```
